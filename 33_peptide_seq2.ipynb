```python
import pandas as pd
import numpy as np

from urllib.request import urlopen
```

```python
tar = pd.read_csv('tar.csv')
```

```python
def cbpId(psid):
    
    html = urlopen('http://www.functionalglycomics.org/glycomics/HServlet?operation=view&sideMenu=no&psId=' + psid)
    page = str(html.read())

    cbpid = page.split('cbpId=')[1]
    cbpid = cbpid.split('&sideMenu=no')[0]
    
    return cbpid
```

```python
cbp_list = []

for psid in tar['primscreen ID']:
    
    try:
        
        cbpid = cbpId(psid)
        
        print(cbpid)
        
        cbp_list.append(cbpid)
    
    except:
        
        print(None)
        
        cbp_list.append(None)
```

```python
tar['CBP ID'] = cbp_list

tar
```

```python
tar.to_csv('tar.csv', index = False)
```
