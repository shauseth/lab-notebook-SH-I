```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

"""plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True"""
```

```python
def plot(x, y, filename = 'untitled', xlabel = 'glycans', ylabel = 'rfus', error = None, size = (20, 5)):
    
    i = np.arange(len(x))
    
    plt.figure(figsize = size)
    plt.bar(i, y, yerr = error, capsize = 3, color = 'black')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(i, x, rotation = 'vertical')
    
    if filename == 'untitled':
        
        plt.show()
    
    else:
        
        plt.savefig(filename + '.eps', format = 'eps', bbox_inches = 'tight', metadata = 'eps')
```
