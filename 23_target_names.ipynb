```python
import pandas as pd
```

```python
tar = pd.read_csv('tar.csv')
```

```python
protein_list = []

for file_id in tar['File ID']:
    
    data = pd.read_excel('cfg_data/' + str(file_id) + '.xls')   
    protein = data.columns[7]
    
    protein_list.append(protein)
    
    print(file_id, protein)
    print('')
```

```python
tar['Target'] = protein_list

tar.to_csv('tar.csv', index = False)
```
