```python
import pandas as pd
import numpy as np

from urllib.request import urlopen
```

```python
tar = pd.read_csv('tar.csv')
```

```python
def Seq(cbpid):
    
    html = urlopen('http://www.functionalglycomics.org/glycomics/GBPServlet?pageType=view&view=view&operationType=view&cbpId=' + cbpid + '&sideMenu=no')
    page = str(html.read())
    
    seq = page.split('align="left" class="rightSideEqualWidth">')[1]
    seq = seq.split('</td>')[0]
    seq = ' '.join(seq.split('&nbsp;'))
    seq = ' '.join(seq.split('<br>'))
    
    return seq
```

```python
seq_list = []

for cbpid in tar['CBP ID']:
    
    if cbpid:
        
        try:

            seq = Seq(cbpid)

            print(seq)

            seq_list.append(seq)

        except:

            print(None)

            seq_list.append(None)
    
    else:
        
        print(None)

        seq_list.append(None)
```

```python
tar['Sequence'] = seq_list

tar
```

```python
tar.to_csv('tar.csv', index = False)
```
