```python
import pandas as pd
import numpy as np
```

```python
input = '20180222-87CLooPA-MR-ppm.txt'
data = pd.read_csv(input, delim_whitespace = True)
data.set_index('Nuc', inplace = True)
```

```python
def ham_dist(s1, s2):

    assert len(s1) == len(s2)

    dist = sum(ch1 != ch2 for ch1, ch2 in zip(s1, s2))

    return dist
```

```python
data = data.sort_values('1-1', ascending = False)
seq1 = data.index[0]
count = list(range(len(data)))[1:]

for x in count:

        seq2 = data.index[x]

        try:

            dist = ham_dist(seq1, seq2)

        except AssertionError:

            dist = None

        if dist == 1:

            print(seq2)
```
