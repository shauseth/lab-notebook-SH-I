```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def glycan_map(protein):
    
    data = pd.read_csv('data/lectins_rfu.csv')

    iupac = data['IUPAC'].values
    rfu = data[protein].values[:600].reshape(25, 24)
    
    plt.figure(figsize = (10, 9))
    plt.imshow(rfu, cmap = 'viridis')
    plt.axis('off')
    
    plt.show()
```

```python
data = pd.read_csv('data/lectins_rfu.csv')

for x in list(data.columns[1:]):
    
    print(x)
    glycan_map(x)
```
