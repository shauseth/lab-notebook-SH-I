```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(val, filename):
    
    plt.figure(figsize = (10, 5))
    plt.plot(val)
    plt.xlabel('glycans')
    plt.ylabel('rfus')
    
    plt.savefig(filename + '.eps', format = 'eps', bbox_inches = 'tight', metadata = 'eps')
```

```python
data = pd.read_excel('Ulex europaeus Agglutinin.xls')
plot(data['Average RFU'], 'fig1')
```

```python
data = data.sort_values('Average RFU')
data = data.reset_index(drop = True)
plot(data['Average RFU'], 'fig2')
```
