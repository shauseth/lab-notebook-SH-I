```python
import pandas as pd
import numpy as np
import string
```

```python
brac = str.maketrans('()', '[]')

def conv(iupac):
    
    iupac = iupac.translate(brac)

    glycan = iupac.split('-')
    
    if len(glycan) == 2:
        
        mono_1 = glycan[0]
        mono_1 = mono_1[:-1] + '(' + mono_1[-1:]
        
        return mono_1 + '1-'

    if len(glycan) == 3:

        mono_1 = glycan[0]
        mono_1 = mono_1[:-2] + '(' + mono_1[-2:]

        mono_2 = glycan[1]
        mono_2 = mono_2[:1] + ')' + mono_2[1:-1] + '(' + mono_2[-1:] + '1'
        
        return mono_1 + '-' + mono_2 + '-'

    else:

        mono_1 = glycan[0]
        mono_1 = mono_1[:-2] + '(' + mono_1[-2:]

        mono_mid = glycan[1:-2]

        for i, x in enumerate(mono_mid):

            x = x[:1] + ')' + x[1:-2] + '(' + x[-2:]

            mono_mid[i] = x

        mono_n = glycan[-2]
        mono_n = mono_n[:1] + ')' + mono_n[1:-1] + '(' + mono_n[-1:] + '1'
        
        return mono_1 + '-' + '-'.join(mono_mid) + '-' + mono_n + '-'
    
def link(iupac):
    
    glycan = iupac.split('-')
    link = glycan[-1]
    
    return link
```

```python
data = pd.read_excel('inconvertible.xlsx')
output = pd.DataFrame()
```

```python
output['IUPAC (CFG)'] = data['IUPAC']
output['IUPAC (GlyTouCan)'] = data['IUPAC'].apply(conv)
output['Linkage'] = data['IUPAC'].apply(link)
```
