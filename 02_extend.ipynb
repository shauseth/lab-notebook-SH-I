```python
import pandas as pd
import os
```

```python
rfu = pd.read_csv('sorted/glycans.csv')
dev = pd.read_csv('sorted/glycans.csv')
```

```python
for filename in os.listdir('sorted/avg_rfu/'):
    
    file = pd.read_csv('sorted/avg_rfu/' + filename)
    
    try:
    
        avg_rfu = file['Average RFU']
        st_dev = file['StDev']

        rfu[filename] = avg_rfu
        dev[filename] = st_dev
    
    except:
        
        pass
```
