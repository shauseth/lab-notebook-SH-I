```python
import pandas as pd
```

```python
iupac = pd.read_excel('IUPAC.xlsx')
mapped = pd.read_excel('mapped.xlsx')
```

```python
iupac.head()
```

```python
mapped.head()
```

```python
id_map = pd.Series(iupac['GlyTouCan ID'].values, index = iupac['IUPAC (Derda Lab)'].values)
```

```python
id_map.head()
```

```python
name_map = pd.Series(iupac['IUPAC (GlyTouCan)'].values, index = iupac['IUPAC (Derda Lab)'].values)
```

```python
name_map.head()
```

```python
mapped['GlyTouCan ID'] = mapped['IUPAC-linker'].map(id_map)
```

```python
mapped.head()
```

```python
mapped['GlyTouCan Name'] = mapped['IUPAC-linker'].map(name_map)
```

```python
mapped
```

```python
mapped.to_csv('mapped.csv', index = False)
```
