```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(x, y, filename, xlabel = 'glycans', ylabel = 'rfus', error = None):
    
    i = np.arange(len(x))
    
    plt.figure(figsize = (5, 5))
    plt.bar(i, y, yerr = error, capsize = 4, color = 'black')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(i, x, rotation = 'vertical')
    
    plt.savefig(filename + '.eps', format = 'eps', bbox_inches = 'tight', metadata = 'eps')
```

```python
data = pd.read_excel('Ulex europaeus Agglutinin.xls')
filt = data[data['Average RFU'] > 5000]
```

```python
plot(filt['Structure on Masterlist'], filt['Average RFU'], 'uea_top_hits', error = filt['StDev'])
```

```python
out = filt[['Structure on Masterlist', 'Average RFU', 'StDev', '% CV']]
```

```python
out.to_csv('UEA_top_hits.csv', index = False)
```
