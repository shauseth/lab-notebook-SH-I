```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
def heat(data, filename = 'untitled', eps = False):
    
    x, y = data

    plt.subplot(1, 2, 1)
    plt.imshow(x, cmap = 'Blues')
    plt.title('output')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(y, cmap = 'Blues')
    plt.title('desired output')
    plt.axis('off')
    
    if filename == 'untitled':
        
        plt.show()

    else:
        
        if eps == True:
            
            plt.savefig(filename + '.eps', format = 'eps', bbox_inches = 'tight', metadata = 'eps')
            
        else:
            
            plt.savefig(filename + '.png', bbox_inches = 'tight', dpi = 5)
```

```python
with open('results.pkl', 'rb') as f:
    
    results = pickle.load(f)
```

```python
x = 6

heat(results[x], 'output_figs/fig' + str(x))
```

```python
# [99, 109, 110] high
# [6, 23, 96, 418] med
```

```python
for x, y in results:
    
    if np.argmax(y) == 0:
        
        heat((x, y))
```
