```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('glycans.csv')
```

```python
excep = ['3S', '4S', '6S', '3P', '4P', '6P', '(Gc)', '-ol', 'Mur', 'Rha']
id_list = []
```

```python
for x in data['IUPAC (CFG)']:
    
    if any(y in x for y in excep):
        
        id_list.append('inconvertible')
    
    else:
        
        id_list.append('convertible')
```

```python
data['action'] = id_list
```
