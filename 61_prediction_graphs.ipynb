```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
```

```python
with open('results.pkl', 'rb') as f:
    
    results = pickle.load(f)
```

```python
out_list = []
d_out_list = []

for out, d_out in results:
    
    out_list.append(np.argmax(out))
    d_out_list.append(np.argmax(d_out))

i = 100

out_arr = np.array(out_list)[i - 100:i]
d_out_arr = np.array(d_out_list)[i - 100:i]

out_arr = np.where(out_arr == 2, 0.1, out_arr)
out_arr = np.where(out_arr == 1, 0.4, out_arr)
out_arr = np.where(out_arr == 0, 0.8, out_arr)

d_out_arr = np.where(d_out_arr == 2, 0.1, d_out_arr)
d_out_arr = np.where(d_out_arr == 1, 0.4, d_out_arr)
d_out_arr = np.where(d_out_arr == 0, 0.8, d_out_arr)
```

```python
fig, ax = plt.subplots(figsize = (20, 5))

index = np.arange(len(out_arr))
bar_width = 0.2

out = plt.bar(index, out_arr, bar_width)
d_out = plt.bar(index + bar_width, d_out_arr, bar_width)
ax.set_xticks(index + bar_width / 2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (20, 5))

index = np.arange(len(out_arr))
bar_width = 0.5

out = plt.scatter(index, out_arr, color = 'royalblue', zorder = 2)
d_out = plt.bar(index, d_out_arr, bar_width, color = 'grey', zorder = 0)

plt.show()
```
