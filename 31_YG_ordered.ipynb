```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cfg = pd.read_csv('worked/Ulex europaeus Agglutinin.csv')
yg = pd.read_csv('UAE_YG.csv')
```

```python
def plot_bar_x(x, y, filename, xlabel = 'glycans', ylabel = 'rfus', error = None):
    
    i = np.arange(len(x))
    
    plt.figure(figsize = (15, 5))
    plt.bar(i, y, yerr = error, capsize = 4)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(i, x, rotation = 'vertical')
    
    plt.savefig(filename + '.png', bbox_inches = 'tight')
```

```python
plot_bar_x(yg['Modific'], yg['total'], 'yg', ylabel = 'enrichment')
```

```python
plot_bar_x(cfg['Structure on Masterlist'], cfg['Average RFU'], 'cfg', error = cfg['StDev'].values)
```
