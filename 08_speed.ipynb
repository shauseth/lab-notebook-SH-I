```python
import pandas as pd
import numpy as np
```

```python
input = '20180530-87CLooOO-JM-ppm.txt'
```

```python
def ham_dist(s1, s2):

    assert len(s1) == len(s2)

    dist = sum(ch1 != ch2 for ch1, ch2 in zip(s1, s2))

    return dist
```

```python
data = pd.read_csv(input, delim_whitespace = True, skiprows = [1])
data.set_index('Nuc', inplace = True)
del data['AA']
```

```python
for x in range(len(data.columns)):

    data.iloc[x] = data.iloc[x].where(data.iloc[x] == data.iloc[x].max(), 0)
```

```python
col_list = []

for x in data:

    col_list.append(x)
```

```python
data = data.sort_values('1-1', ascending = False)

seq1 = data.index[0]

count = list(range(len(data)))[1:]
```

```python
for x in count:

    seq2 = data.index[x]
        
    try:

        dist = ham_dist(seq1, seq2)

    except AssertionError:

        pass
    
    if dist == 1:
        
        data.loc[0] = np.array(data.iloc[0]) + np.array(data.iloc[x])
        data = data.drop([seq2])
        
        del count[-1]
        
        print(data.head())
```
