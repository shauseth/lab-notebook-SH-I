```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lectins = ['ACG', 'CTB', 'DC_SIGN', 'Gal4', 'LCA', 'MAL', 'NPL', 'PSA', 'SBA', 'SNL', 'UEA']
```

```python
l = 0
n = 305

data = pd.read_excel('figures/cfg/CFG_' + lectins[l] + '.xls')
data = data.iloc[n:n + 100, :]
```

```python
def plot(data, save = False):
    
    x, y, e = data['Structure on Masterlist'], data['Average RFU'], data['StDev']

    i = np.arange(len(x))
    
    fig, ax = plt.subplots(figsize = (32, 10))

    ax.bar(i, y, yerr = e, capsize = 3, color = 'black')
    plt.sca(ax)
    plt.xlabel('Glycans')
    plt.ylabel('Average RFU')
    plt.xticks([], [])
    
    if save:
        
        plt.savefig(fname = 'fig3.pdf', bbox_inches = 'tight')
    
    else:
        
        plt.show()
```

```python
plot(data, save = True)
```
